```python
from dataclasses import dataclass
from typing import List
from datetime import date

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

##### DEFINE THE CRITIQUE CLASS

```python
@dataclass
class Critique:
    critique_id: str
    painting_id: str
    critic_name: str
    title: str
    text: str
    published_at: date
    label: int   # 1 = expert-like, 0 = not expert-like
```

##### LABELED CRITIQUE EXAMPLES

##### CONVERT THE CRITIQUES INTO A DATAFRAME

```python
df = pd.DataFrame([c.__dict__ for c in critiques])

df[["critique_id", "critic_name", "title", "label"]]
```

##### BUILD A CLASSICAL ML PIPELINE : BUILD TF-IDF + LOGISTIC REGRESSION PIPELINE

```python
model = Pipeline([
    ("tfidf", TfidfVectorizer(
        ngram_range=(1, 2),
        lowercase=True,
        min_df=1,
        stop_words="english"
    )),
    ("classifier", LogisticRegression())
])
```

##### TRAIN THE MODEL

```python
X_train = df["text"]
y_train = df["label"]

model.fit(X_train, y_train)
```

##### CLASSIFY A NEW STUDENT CRITIQUE TO SEE HOW EXPERT-LIKE IT SOUNDS

```python
new_student_critique = """
The painting uses contrast and light to guide the viewer's attention.
The placement of the figures creates depth, while the darker areas add emotional tension.
"""

prediction = model.predict([new_student_critique])[0]
probability = model.predict_proba([new_student_critique])[0][1]

print("Predicted label:", prediction)
print("Probability of expert-like critique:", round(probability, 3))
```

##### INSPECT LEARNED FEATURE WEIGHTS

```python
tfidf = model.named_steps["tfidf"]
classifier = model.named_steps["classifier"]

feature_names = tfidf.get_feature_names_out()
coefficients = classifier.coef_[0]

coef_df = pd.DataFrame({
    "feature": feature_names,
    "coefficient": coefficients
})

top_expert_features = coef_df.sort_values(
    by="coefficient",
    ascending=False
).head(15)

top_non_expert_features = coef_df.sort_values(
    by="coefficient",
    ascending=True
).head(15)

print("Top features associated with expert-like critiques:")
display(top_expert_features)

print("Top features associated with less expert-like critiques:")
display(top_non_expert_features)
```
